```python
ro=df[df['batter']=='RG Sharma']
sum(ro.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False))
```

```python
def rusnscored(batsman):
    mask=df['batter']==batsman
    newdf=df[mask] 
    return newdf.groupby('bowling_team')['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python
rusnscored('MS Dhoni')
```

```python
death_overs=df[df['over']>15]
```

```python
a=death_overs.groupby('batter')['batsman_runs'].sum()
b=a[a>200]
batsman_list=b.index.to_list()
batsman_list
```

```python
final_list=df[df['batter'].isin(batsman_list)]
```

```python
runs=final_list.groupby('batter')['batsman_runs'].sum()
runs
```

```python
balls=final_list.groupby('batter')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
sr.sort_values(ascending=False).head(30
                                    )
```
